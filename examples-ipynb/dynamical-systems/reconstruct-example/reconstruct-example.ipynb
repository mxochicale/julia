```julia
## 3D continuous chaotic system
### https://juliadynamics.github.io/DynamicalSystems.jl/latest/embedding/reconstruction/
using DynamicalSystems, PyPlot
ds = Systems.gissinger(ones(3))
data = trajectory(ds, 1000.0, dt = 0.05)
xyz = columns(data)
```

```julia
figure(figsize = (12,10))
k = 1
for i in 1:3
    for τ in [5, 100, 200]
        R = reconstruct(xyz[i], 1, τ)
        ax = subplot(3,3,k)
        plot(R[:, 1], R[:, 2], color = "C$(k-1)", lw = 0.8)
        title("var = $i, τ = $τ")
        global k+=1
    end
end
```
